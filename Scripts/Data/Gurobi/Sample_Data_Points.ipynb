# Notebook to Sample from Data Set

```python
# At random, need distribution of season, time but not equal

import pandas as pd
import numpy as np
import random

# Read Files
busses = pd.read_excel('Busses_Timeseries_Table_Set2022_ExcelEdit.xlsx')
edges = pd.read_excel('Edges_Timeseries_Table_Set2022_ExcelEdit.xlsx')
```

```python
timestamps = edges['datetime_beginning_utc'].unique()
times = timestamps.size

sampleSize = 1000
sample = random.sample(list(timestamps), sampleSize)

bussesSample = busses.loc[busses['datetime_beginning_utc'].isin(sample)]
bussesSample = bussesSample.sort_values(by=['datetime_beginning_utc', 'bus_id'], ascending=True)

edgesSample = edges.loc[edges['datetime_beginning_utc'].isin(sample)]
edgesSample = edgesSample.sort_values(by=['datetime_beginning_utc', 'From Number', 'To Number'], ascending=True)
```

```python
bussesSample.to_csv("Busses_Timeseries_Table_Set2022_Sample.csv")
edgesSample.to_csv("Edges_Timeseries_Table_Set2022_Sample.csv")
```
